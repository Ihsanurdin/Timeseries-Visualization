```python
import pandas as pd
import matplotlib.pyplot as plt

datafr = pd.read_csv("/content/drive/My Drive/Saas/data_balita.csv", encoding='unicode_escape')

filterdata = datafr[datafr['Umur (bulan)'] > 36]

T = filterdata.head(30)['Tinggi Badan (cm)']
S = filterdata.head(30)['Status Gizi']

fig, ax = plt.subplots()
ax.bar(S, T, color='#343F56')

ax.set_title('Status Gizi Balita (Umur > 36 bulan) Menurut Tinggi Badan')
ax.set_xlabel('Status Gizi')
ax.set_ylabel('Tinggi Badan (cm)')

plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

# Melakukan filter untuk anak-anak dengan usia lebih dari 36 bulan
filterdata = df[df['Umur (bulan)'] > 36]

# Menggunakan groupby untuk menghitung jumlah data untuk setiap kategori status gizi
status_gizi_count = filterdata.groupby('Status Gizi').size()

# Membuat plot bar horizontal
plt.figure(figsize=(10, 6))
status_gizi_count.plot(kind='barh', color=sns.color_palette('Dark2'))
plt.title('Jumlah Anak Berdasarkan Status Gizi (Usia >36 Bulan)')
plt.xlabel('Jumlah Anak')
plt.ylabel('Status Gizi')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Melakukan filter untuk anak-anak dengan usia lebih dari 36 bulan
filterdata = df[df['Umur (bulan)'] > 36]

# Mendefinisikan data untuk plotting
umur = filterdata['Umur (bulan)']
status_gizi = filterdata['Status Gizi']

# Membuat scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(umur, status_gizi, color='blue', alpha=0.5)
plt.title('Scatter Plot Umur (bulan) vs Status Gizi (Usia >36 Bulan)')
plt.xlabel('Umur (bulan)')
plt.ylabel('Status Gizi')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
